### 2. Overfitting은 무엇을 의미하며 해결방법은 무엇인지 기술하시오.

### answer
- **Overfitting(과적합)**이란, 학습(train) 데이터를 과하게 학습(심지어, 노이즈도 학습)하여 학습데이터에 한정하여 성능을 100% 가깝게 보이나, 테스트(test, unseen) 데이터에 대해서는 성능이 현저히 떨어지는 것(정확성 하락)을 말한다.
- 흔히, 과적합된 모델은 복잡하다고 말할 수 있다., 설명하기도 어렵다.
- 예를 들어, **서포트 벡터 머신(svm)** 경우에, *C와 *Gamma의 값을 상당히 크게 잡아서(오버피팅 발생) decision boundary를 구불구불하고 복잡하게 만들어 새로운 데이터를 모델에 적용했을때 제대로 성능을 발휘하지 못하는 경우가 있다.
    - *C : Controls tradeoff detween smooth decision boundary and classfying training points correctly(오차를 허용하는 정도)
        1. c의 값이 커지면, margin이 줄어들어 오차율이 줄어든다.-> 더욱 구불구불해짐
        2. c의 값이 작아지면, margin이 늘어난다. -> 선형에 가까워짐
    - *Gamma: Defines how far the influence of a single training point reaches(decision boundary의 곡률 정도)
        1. gamma의 값이 커지면, reach가 가깝기 때문에 경계선에 놓인 포인트들에 영향을 크게 받는다 -> 더욱 구불구불해짐
        2. gamma의 값이 작아지면, reach가 멀기 때문에 가까이 있는 포인트의 영향이 상대적으로 떨어진다
##### →따라서, 서포트 벡터 머신(svm) 경우에, grid search를 이용하여 하이퍼 파라미터인 C와 Gamma의 best parameter(최적의 매개변수)를 찾아낸다.

### 3. 표준화와 정규화에 대해 조사하고 왜 필요한지 머신러닝 기법 중 data scaling 에 민감한 분석법이 있는지 조사하세요.

### Data scaling
- 데이터에는 다양한 특성(feature)가 포함되어 있다. 하지만 이 특성들의 단위와 범위는 모두 다르기 때문에 제대로 비교를 할 수 없다.→ 실제로, 모델링 후 accuracy에서 확연한 차이가 난다.
    - ex) 인구 수(명), 소득 금액(원), 연령(세) 등
- 따라서, 데이터 분석의 정확한 비교를 위해 **특성들의 단위를 무시**하거나 **특성 값의 범위를 비슷**하게 만들어주는 것이 **Data scaling**이다.
    1. **정규화(normalization)** : 특성 값의 범위를 [0,1]로 축소시킨다.
        - 최소값은 0, 최대값은 1로 비율의 값을 갖게 된다.
    2. **표준화(standardization)** : 특성 값이 정규분포(0,1)을 따르도록 값을 대체한다.
        - 특성이 정규분포를 따른다고 가정하고 평균이 0, 표준편차가 1을 따르게 하여 모든 특성이 같은 크기를 가지게 된다.

### When we should use 'scaling'?

#### 1. 서포트벡터머신(SVM)
- 서포트벡터 머신은 서포트벡터와 분류기(hyperplane) 사이 거리인 마진(margin)을 최대로 만들어주는 분류기를 찾는 알고리즘이다. 따라서, **큰 값을 가지는 변수가 거리 계산을 할 때 영향력을 많이 미치게 된다.**

#### 2. 주성분 분석(PCA)
- 주성분 분석에서 높은 분산과 넓은 범위를 가지는 변수는 낮은 분산을 가지는 변수보다 주성분에서 큰 회귀계수를 가지게 된다. **즉, 중요하지 않은 변수도 범위가 넓으면 주성분에서 중요한 변수로 간주된다**는 것이다. 이는 모델링을 과정에서 가장 치명적이다.

#### 3. 군집(clustering)
- 군집분석은 **거리-기반** 알고리즘으로 거리 측도(ex. 유클리디안, 맨하튼 등)를 이용하여 관측값 사이에서 유사성을 찾아 군집을 형성한다. 따라서, 넓은 범위를 가지는 변수는 군집에서 더 큰 영향력을 가지게 된다.

#### 4. k-최근접 이웃(knn)
- k-최근접 이웃은 **거리-기반** 알고리즘으로 새로운 관측값의 주변에 있는 k개의 이웃(데이터)를 이용하여 유사성 측도에 기반해 관측값을 분류한다. 따라서, 유사성 측도에 모든 변수가 동일하게 기여할 수 있도록 스케일링을 수행해야한다.

#### 5. 회귀
- 예측을 위한 회귀 모델링이 목적이라면 표준화를 수행할 필요가 없지만, 변수의 중요도나 다른 회귀계수들과 비교를 하는 것이 목적이라면 표준화를 수행하여야 한다.

#### 6. 릿지-라쏘
- 릿지-라쏘 모형은 회귀계수에 벌점을 가하는 방법으로 다중공선성과 과적합을 해결하기 위해 사용된다. 회귀계수의 크기를 벌점으로 사용하기 때문에 상대적으로 큰 분산을 가지는 변수가 회귀계수가 클 것이고 그 회귀계수를 벌점으로 사용하게 되면 작은 분산을 가지는 변수들보다 회귀계수를 더 **축소시켜 최종적인 회귀계수는 작아지게 될 것이다.**
- 정리하면 릿지, 라쏘는 회귀계수가 단위에 영향을 미치기 때문에 표준화를 꼭 수행해야한다.

### When we don't need to use 'scaling'?
- 로지스틱 회귀분석
- 트리 기반 모델(의사결정나무, 랜덤 포레스트 , 그래디언트 부스팅)
